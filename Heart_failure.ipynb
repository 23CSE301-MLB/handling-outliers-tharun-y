```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['ejection_fraction'] = scaler.fit_transform(df[['ejection_fraction']])
df['serum_creatinine'] = scaler.fit_transform(df[['serum_creatinine']])
```

```python
df['creatinine_phosphokinase'] = scaler.fit_transform(df[['creatinine_phosphokinase']])
```

```python
sns.boxplot(x='creatinine_phosphokinase' , data = df)
```

```python
Q1 = df['creatinine_phosphokinase'].quantile(0.25)
Q3 = df['creatinine_phosphokinase'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
outlier_iqr = df[(df['creatinine_phosphokinase'] < lower_bound )| (df['creatinine_phosphokinase'] > upper_bound)]
print(outlier_iqr.shape)
```

```python
import numpy as np
df['creatinine_phosphokinase'] = np.log(df['creatinine_phosphokinase']+1)
```

```python
df
```

```python
median = df['creatinine_phosphokinase'].median()

df.loc[(df['creatinine_phosphokinase'] < lower_bound) | 
       (df['creatinine_phosphokinase'] > upper_bound), 'creatinine_phosphokinase'] = median
```
